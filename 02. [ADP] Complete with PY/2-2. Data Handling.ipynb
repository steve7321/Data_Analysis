```python
import pandas as pd
students = pd.DataFrame({'이름':['A','B','C','D','E','F'],'국어':[70,85,None,100,None,85],'수학':[65,100,80,95,None,70]})
students[students['이름']=='A']
```

```python
students[(students['국어']>80)&(students['수학']>80)]
```

```python
students.loc[6,'이름':'수학'] = ['G',50,60]
students.loc[(students['국어']>=80)&(students['수학']>=70),'합격']='Pass'
students.loc[students['합격']!='Pass','합격']='Fail'
students
```

```python
import numpy as np

score_list = [(students['국어']>=90),(students['국어']>=80)&(students['국어']<90),(students['국어']>=70)&(students['국어']<80)]
grade_list = ['A','B','C']
students['점수']=np.select(score_list,grade_list,default='F')
students
```

```python
students.isna()
students.notna()
students.isna().sum()
students.isna().sum(1)  #행 합계를 나타냄 
```

```python
students.dropna()
students.dropna(thresh=4)
students.dropna(how='any')
students.dropna(subset='수학')
```

```python
health = pd.DataFrame({'연도':[2017,2018,2019,2020,2021,2022],'키':[160,162,165,None,None,166], \
                      '몸무게':[53,52,None,50,51,54],'시력':[1.2,None,1.2,1.2,1.1,0.8], \
                      '병결':[None,None,None,2,None,1]})
health.fillna(0)
health.fillna(health.mean())

health['병결']=health['병결'].fillna(0)
health['몸무게']=health['몸무게'].fillna(health['몸무게'].mean())

#health.fillna(method='pad',inplace=True)
health.ffill(inplace=True)
health
```
