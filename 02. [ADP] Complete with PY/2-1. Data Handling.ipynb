```python
score = pd.DataFrame([[100, 75, 90],[80, 90, 95]],columns=['국어','수학','영어'],index=['장화','홍련'])
score['국어'] = score['국어'] - 5
score.loc['홍련','영어']=100
score
```

```python
new_students = pd.DataFrame ({'국어':[70,85],'영어':[95,65],'수학':[65,100]},index=['콩쥐','팥쥐'])
score=pd.concat([score,new_students])
score
```

```python
new_student1=pd.Series({'국어':85, '수학':55, '영어':95},name='해님')
score = pd.concat([score, new_student1.to_frame().T])
score
```

```python
science=[80, 70, 90, 85, 75]
score['과학']=science
score['학년']=1
score
```

```python
score['과학']=score['과학']+5
score['총점']=score['국어']+score['수학']+score['영어']+score['과학']
score
```

```python
score.drop(['장화','해님'])
score.drop(['국어','영어'],axis=1)
score.drop(columns=['국어','영어'])
```
