```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

wine_load = load_wine()
wine = pd.DataFrame (wine_load.data, columns=wine_load.feature_names)
wine['class']=wine_load.target
wine['class']=wine['class'].map({0:'class 0',1:'class 1',2:'class 2'})

wine_type = wine['class'].value_counts(sort=False)
wine_type
```

```python
plt.figure(figsize=(6,4))
plt.bar(wine_type.index, wine_type.values, width=0.8, bottom=None, align='center')
plt.title('number of wine classes')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.barh(wine_type.index, wine_type.values, height=0.8, left=None, align='center')
plt.title('number of wine classes')
plt.show()
```

```python
plt.figure(figsize=(6,4))
#plt.hist('alcohol', bins=8, range=(11,15), color='purple', data=wine)
plt.hist(wine['alcohol'], bins=8, range=(11,15), color='purple')
plt.title('wine alcohol')
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

IRIS=load_iris()
iris=pd.DataFrame (IRIS.data, columns=IRIS.feature_names)
iris['class'] = IRIS.target
iris['class'] = iris['class'].map({0:'setosa',1:'versicolor',2:'virginica'})

plt.boxplot(iris.drop(columns='class'), tick_labels=iris.columns[0:4])
plt.show()
```

```python
plt.boxplot(iris['sepal width (cm)'], whis=1.5)
plt.title('sepal width (cm)')
plt.show()
```

```python
iris[['sepal width (cm)','class']].boxplot(by='class')
plt.title('sepal width (cm)')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='class',y='sepal width (cm)',data=iris)
plt.show()
```

```python
plt.title('iris scatter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(iris['sepal length (cm)'],iris['sepal width (cm)'],alpha=0.5)
plt.show()
```

```python
sns.scatterplot(x='sepal length (cm)',y='sepal width (cm)',data=iris,hue='class',style='class')
plt.show()
```
